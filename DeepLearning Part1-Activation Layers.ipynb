```python
import numpy as np

# Sigmoid activation function
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# Example of applying the sigmoid activation to a layer's output
def sigmoid_layer(inputs):
    return sigmoid(inputs)

# Sample input (e.g., output from a dense layer)
inputs = np.array([-25,0.5, -1.0, 2.0,50,100])

# Apply the sigmoid activation layer
outputs = sigmoid_layer(inputs)

print("Input:", inputs)
print("Sigmoid Activation Output:", outputs)
```

```python
import numpy as np

# Linear activation function
def linear(x):
    return x  # Linear activation simply returns the input without modification

# Example of applying the linear activation to a layer's output
def linear_layer(inputs):
    return linear(inputs)

# Sample input (e.g., output from a dense layer)
inputs = np.array([0.5, -1.0, 2.0, -3.5, 1.5])

# Apply the linear activation layer
outputs = linear_layer(inputs)

print("Input:", inputs)
print("Linear Activation Output:", outputs)
```

```python
import numpy as np

# ReLU activation function
def relu(x):
    return np.maximum(0, x)

# Example of applying the ReLU activation to a layer's output
def relu_layer(inputs):
    return relu(inputs)

# Sample input (e.g., output from a dense layer)
inputs = np.array([-25,0.5, -1.0, 2.0,50,100])

# Apply the ReLU activation layer
outputs = relu_layer(inputs)

print("Input:", inputs)
print("ReLU Activation Output:", outputs)
```

```python
import numpy as np

# Tanh activation function
def tanh(x):
    return np.tanh(x)

# Example of applying the Tanh activation to a layer's output
def tanh_layer(inputs):
    return tanh(inputs)

# Sample input (e.g., output from a dense layer)
inputs = np.array([-123,0.5, -1.0, 2.0, -3.5, 1.5,232])

# Apply the Tanh activation layer
outputs = tanh_layer(inputs)

print("Input:", inputs)
print("Tanh Activation Output:", outputs)
```

```python
import numpy as np

# Softmax activation function
def softmax(x):
    # Subtract the max for numerical stability (prevents overflow issues)
    exp_x = np.exp(x - np.max(x))
    return exp_x / np.sum(exp_x)

# Example of applying the Softmax activation to a layer's output
def softmax_layer(inputs):
    return softmax(inputs)

# Sample input (e.g., output from a dense layer)
inputs = np.array([1.0, 2.0, 3.0,23,45])
inputs1 = np.array([23.0, 2.0, 43.0,23,5])

# Apply the Softmax activation layer
outputs = softmax_layer(inputs1)

print("Input:", inputs1)
print("Softmax Activation Output:", outputs)
```

```python
import numpy as np

# ELU activation function
def elu(x, alpha=1.0):
    return np.where(x >= 0, x, alpha * (np.exp(x) - 1))

# Example of applying the ELU activation to a layer's output
def elu_layer(inputs, alpha=1.0):
    return elu(inputs, alpha)

# Sample input (e.g., output from a dense layer)
inputs = np.array([0.5, -1.0, 2.0, -3.5, 1.5])

# Apply the ELU activation layer
outputs = elu_layer(inputs)

print("Input:", inputs)
print("ELU Activation Output:", outputs)
```

```python
import numpy as np

# Leaky ReLU activation function
def leaky_relu(x, alpha=0.1):
    return np.where(x >= 0, x, alpha * x)

# Example of applying the Leaky ReLU activation to a layer's output
def leaky_relu_layer(inputs, alpha=0.1):
    return leaky_relu(inputs, alpha)

# Sample input (e.g., output from a dense layer)
inputs = np.array([-100,0.5, -1.0, 2.0, -3.5, 1.5,200])

# Apply the Leaky ReLU activation layer
outputs = leaky_relu_layer(inputs)

print("Input:", inputs)
print("Leaky ReLU Activation Output:", outputs)
```

```python
import numpy as np

# GELU activation function
def gelu(x):
    return 0.5 * x * (1 + np.tanh(np.sqrt(2 / np.pi) * (x + 0.044715 * x**3)))

# Example of applying the GELU activation to a layer's output
def gelu_layer(inputs):
    return gelu(inputs)

# Sample input (e.g., output from a dense layer)
inputs = np.array([-25,-10,0.5, -1.0, 2.0, -3.5, 1.5,25,50])

# Apply the GELU activation layer
outputs = gelu_layer(inputs)

print("Input:", inputs)
print("GELU Activation Output:", outputs)
```
